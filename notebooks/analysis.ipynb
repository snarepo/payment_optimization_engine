```python
# This is a Jupyter notebook pseudocode snippet:

import sqlite3
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

conn = sqlite3.connect("../data/experiment.db")
df = pd.read_sql_query("""
    SELECT a.group_name, p.earnings, d.churned
    FROM assignments a
    JOIN payouts p ON a.driver_id = p.driver_id
    JOIN drivers d ON d.driver_id = a.driver_id
""", conn)

sns.boxplot(x="group_name", y="earnings", data=df)
plt.title("Earnings Distribution by Group")
plt.show()

sns.barplot(x="group_name", y="churned", data=df)
plt.title("Churn Rate by Group")
plt.show()
```
